```python
#  import libraries and load model
import torch
import cv2
from pathlib import Path

# Path to  custom model
model_path = Path('/content/drive/My Drive/AI_5/Video_Detection/model/iteration_2_best.pt')

# Load YOLOv5 custom model (from local clone)
model = torch.hub.load('/content/yolov5', 'custom', path=str(model_path), source='local')
#  adjust confidence and IoU thresholds
model.conf = 0.25  # minimum confidence
model.iou  = 0.45  # NMS IoU threshold
```

```python
#  run graffiti detection on each video and save annotated outputs
import os

# Folders
video_dir  = Path('/content/drive/My Drive/AI_5/Video_Detection/videos')
output_dir = Path('/content/drive/My Drive/AI_5/Video_Detection/output_videos')
output_dir.mkdir(parents=True, exist_ok=True)

# Process each video file
for vid in sorted(video_dir.iterdir()):
    if vid.suffix.lower() not in ('.mp4', '.avi', '.mov', '.mkv'):
        continue

    cap = cv2.VideoCapture(str(vid))
    width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
    height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
    fps    = cap.get(cv2.CAP_PROP_FPS) or 30.0

    # Prepare VideoWriter
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out_path = output_dir / vid.name
    out = cv2.VideoWriter(str(out_path), fourcc, fps, (width, height))

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # run inference (returns a Results object)
        results = model(frame)

        # render() draws boxes & labels onto the frame array
        annotated_frame = results.render()[0]

        # write annotated frame
        out.write(annotated_frame)

    cap.release()
    out.release()
    print(f"Saved annotated video to: {out_path}")
```
